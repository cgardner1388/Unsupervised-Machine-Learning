```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
### Data Preparation
```

```python
file_path = Path("crypto_data.csv")
df = pd.read_csv(file_path)
df.head()
```

```python
df.dtypes
```

```python
#Filter the data to only true values
trading_df = df.loc[df['IsTrading'] == True]
trading_df.head()
```

```python
#Show that only true values are in the table
trading_df['IsTrading'].value_counts()
```

```python
#Drop the IsTrading column
crypto_trading_df = trading_df.drop(columns=['IsTrading'])
crypto_trading_df.head()
```

```python
#Remove the null values from the dataframe
crypto_trading_dropna_df = crypto_trading_df.dropna(axis=0, how="any", thresh=None, subset=None, inplace=False)
crypto_trading_dropna_df
```

```python
#48 Rows with zero
crypto_trading_dropna_df['TotalCoinSupply'].value_counts()
```

```python
#Remove anything with a zero in the TotalCoinSupply column
cleaned_crypto_df = crypto_trading_dropna_df.loc[crypto_trading_dropna_df['TotalCoinSupply'] > '0']
cleaned_crypto_df
```

```python
#Remove the column CoinName
final_cleaned_crypto_df= cleaned_crypto_df.drop(columns=['CoinName'])
final_cleaned_crypto_df
```

```python
###convert the remaining features with text values, `Algorithm` and `ProofType`, into numerical data. To accomplish this task, use Pandas to create dummy variables. Examine the number of rows and columns of your dataset now. How did they change?
###Standardize your dataset so that columns that contain larger values do not unduly influence the outcome.
cleaned_crypto_ml_df = pd.get_dummies(data=final_cleaned_crypto_df, columns=['Algorithm', 'ProofType'])
cleaned_crypto_ml_df.head()
```

```python
final_cleaned_crypto_ml_df = cleaned_crypto_ml_df.set_index('Unnamed: 0')
final_cleaned_crypto_ml_df
```

```python
###Get dummies, scale, PCA, PCA needs to be on the scale data
```

```python
crypto_scaled_df = StandardScaler().fit_transform(final_cleaned_crypto_ml_df)
```

```python
### Dimensionality Reduction
### Use PCA
```

```python
crypto_scaled_df.shape
```

```python
# Initialize PCA model
pca = PCA(n_components=.90)

```

```python
# Get two principal components for the data.
crypto_pca = pca.fit_transform(crypto_scaled_df)
crypto_pca.shape
```

```python
# Transform PCA data to a DataFrame
crypto_pca_df = pd.DataFrame(crypto_pca)
crypto_pca_df.head()
```

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
```

```python
# Initialize t-SNE model
tsne = TSNE(learning_rate=35)
```

```python
# Reduce dimensions
tsne_features = tsne.fit_transform(crypto_pca_df)
```

```python
# Prepare to plot the dataset

# The first column of transformed features
crypto_pca_df['x'] = tsne_features[:,0]

# The second column of transformed features
crypto_pca_df['y'] = tsne_features[:,1]
```
